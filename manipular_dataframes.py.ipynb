```python
df_auth = ieee[['author']].replace(";", " and", regex=True)
df_auth['author'].replace('', np.nan, inplace=True)
df_auth.dropna(subset=['author'], inplace=True)
#df_auth
df_auth.head(10)
```

```python
#função para inverter a ordem do nome/sobrenome e  abreviar os nomes escritos por extenso
def organize(x):
    ret = []
    for s in x:
        author = ""
        dots = s.count('.')
        if dots == 0:
            l = s.split(' ')
            author += l.pop().strip() + ","
            for i in l:
                author += " " + l.pop(0).strip()[0] + "."
        else:
            l = s.split('.')
            author += l.pop().strip() + ","    
            for i in range(0,dots):
                author += " " + l.pop(0).strip() + "."    
        ret.append(author)
    return ret
```

```python
author_series = df_auth.author.str.split(' and ')
#.map(lambda x : [s.strip() for s in x])
#.map(Counter).sum())
author_series = author_series.apply(organize)
#author_series.head(2)
series_bonita = author_series.map(lambda x : " and ".join(x))
series_bonita.head(2)
```

```python
#update dataframe with correct values
data.loc[source_ieee,['author']] = series_bonita
#data.loc[source_ieee]
data.tail(5)
```

```python
data = pd.ExcelFile("dataframe_only_useful.xls").parse(0, index_col=0)
#data.head(10)
data.columns
```

```python
source_acm = data.source.str.contains('ACM')
acm = data.loc[source_acm]
data.loc[source_acm,['document_type']] = acm['document_type'].apply(lambda x : str(x).replace('inproceedings','conference'))
acm.head(2)
```

```python
acm[['author']].head(10)
```

```python
#função para inverter a ordem do nome/sobrenome e  abreviar os nomes escritos por extenso
def organize(x):
    ret = []
    debbugging = True
    if debbugging: print('INICIO: ', x)
    for s in x:
        #if debbugging: print('DENTRO FOR: ', s)
        author = ""
        dots = s.count('.')
        l = s.split(',')
        if dots == 0: #nomes na ordem correta porém por extenso, precisa abreviar
            if debbugging: print(l[0], '   -   ', l)
            author += l.pop(0).strip() + ","
            if len(l) > 0:
                l2 = l[0].strip().split(' ')
                for i in l2:
                    author += " " + l2.pop(0).strip()[0] + "."
        elif len(l[1]) == l[1].count('.')*2 + l[1].count(' '): # nomes já estão abreviados e na ordem correta, só manter
            author += s
        else: # algum caso bizarro
            author += l[0].strip() + ','
            l2 = l[1].strip().split(' ')
            #print(l2)
            for s in l2:
                if s.count('.') == 1:
                    author += ' ' + s.strip()
                else:
                    author += ' ' + s.strip()[0] + '.'
        if debbugging: print("RET: ", author)
        ret.append(author)
    return ret
```

```python
df_auth = acm[['author']]
author_series = df_auth.author.str.split(' and ')
#.map(lambda x : [s.strip() for s in x])
#.map(Counter).sum())
author_series = author_series.apply(organize)
#author_series.head(2)
series_better = author_series.map(lambda x : " and ".join(x))
series_better
```

```python
#update dataframe with correct values
data.loc[source_acm,['author']] = series_better
#data.loc[source_ieee]
data.tail(15)
```

```python
import re

def insert_space(string, index):
    return string[:index] + ' ' + string[index:]

def add_space(string):
    exp = re.compile('[.]\S')
    result = exp.findall(string)
    for r in result:
        string = string.replace(r, insert_space(r, r.index('.')+1))
    return string


data['author'] = data['author'].apply(lambda x : add_space(x))
```
